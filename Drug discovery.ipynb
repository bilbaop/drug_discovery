```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

Search for Target Protein

```python
target = new_client.target
target_query = target.search('aromatase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

```python
#Here, we will retrieve only bioactivity data for aromatase (CHEMBL1978) that are reported as IC values in nM (nanomolar) unit. 
# Half-maximal inhibitory concentration (IC50) is the most widely used and informative measure of a drug's efficacy. 
# It indicates how much drug is needed to inhibit a biological process by half, thus providing a measure of potency of an antagonist drug in pharmacological research.

activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
#delete compounds that not include the standard value:
df2 = df[df.standard_value.notna()]
df2
```

```python
#Data pre-processing in inactive, active and intermidiate:
bioactivity_class=[]
for i in df2.standard_value:
    if float (i) >= 10000:
        bioactivity_class.append("inactive")
    elif float (i) <= 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermidiate")
```

```python
selection= ['molecule_chembl_id' , 'canonical_smiles', 'standard_value']
df3=df2[selection]
pd.concat([df3,pd.Series(bioactivity_class)],axis=1)
```

Standard: potency of the drug, the lower the number the better the potency. the higher, the worse. we want to find as low as possible. inhibitory concentration at 50% you will need the lower amount of the drug.
